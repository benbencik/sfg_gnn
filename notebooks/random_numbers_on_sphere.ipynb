```python
import numpy as np
import matplotlib.pyplot as plt
```

***

## 2 dimensions

```python
NUM_POINTS = 1000  # number of generated points
RAD = 1.           # vector radius
```

```python
Theta = np.random.uniform(0., 2.*np.pi, NUM_POINTS)  # random numbers between 0 and 2*pi
X, Y = RAD*np.cos(Theta), RAD*np.sin(Theta)  # vector coordinates
```

```python
fig, ax = plt.subplots(figsize=(10,10))

origin = np.array([[0]*NUM_POINTS,[0]*NUM_POINTS])
ax.quiver(*origin, X, Y, scale=1, scale_units="xy", angles="xy", color="tab:red")

ax.set_xlim((-1.2*RAD, 1.2*RAD))
ax.set_ylim((-1.2*RAD, 1.2*RAD))

ax.set_xlabel("x")
ax.set_ylabel("y")

ax.set_title(r"{:d} random vectors with uniformly distributed angle $\theta$".format(NUM_POINTS))

plt.show()
```

***

## 3 dimensions

```python
NUM_POINTS_3D = 10000  # number of points in 3D
```

```python
Theta = np.random.uniform(0., np.pi, NUM_POINTS_3D)  # random numbers between 0 and pi
Phi = np.random.uniform(0., 2.*np.pi, NUM_POINTS_3D) # random numbers between 0 and 2*pi

X = RAD*np.cos(Phi)*np.sin(Theta)
Y = RAD*np.sin(Phi)*np.sin(Theta)
Z = RAD*np.cos(Theta)
```

```python
# 3D view
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection="3d")

ax.scatter(X, Y, Z, alpha=0.2, c=Z)

ax.set_xlim((-1.2*RAD, 1.2*RAD))
ax.set_ylim((-1.2*RAD, 1.2*RAD))
ax.set_zlim((-1.2*RAD, 1.2*RAD))

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

ax.set_title(r"{:d} random vectors with uniformly distributed angles $\theta$ and $\phi$".format(NUM_POINTS_3D))

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11,5))

# top view
ax1.scatter(X, Y, c=Z, alpha=0.2, s=10)

ax1.set_xlim((-1.2*RAD, 1.2*RAD))
ax1.set_ylim((-1.2*RAD, 1.2*RAD))

ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax1.set_title(r"Top view")

# side view
ax2.scatter(X, Z, c=Z, alpha=0.2, s=10)

ax2.set_xlim((-1.2*RAD, 1.2*RAD))
ax2.set_ylim((-1.2*RAD, 1.2*RAD))

ax2.set_xlabel("x")
ax2.set_ylabel("z")

ax2.set_title(r"Side view")

plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,3))

# x-components
ax1.hist(X, bins=np.linspace(-1, 1, 100), color="tab:blue")

ax1.set_xlabel("x value")
ax1.set_ylabel("number of occurences")

# y-components
ax2.hist(Y, bins=np.linspace(-1, 1, 100), color="tab:blue")

ax2.set_xlabel("y value")
ax2.set_ylabel("number of occurences")

# z-components
ax3.hist(Z, bins=np.linspace(-1, 1, 100), color="tab:blue")

ax3.set_xlabel("z value")
ax3.set_ylabel("number of occurences")

# ax.set_title("histogram")

plt.show()
```

### Uniformly distributed vectors

```python
Lambda = np.arccos(2.*np.random.uniform(0., 1., NUM_POINTS_3D) - 1.)
Phi = np.random.uniform(0., 2.*np.pi, NUM_POINTS_3D) # random numbers between 0 and 2*pi

X = RAD*np.cos(Phi)*np.sin(Lambda)
Y = RAD*np.sin(Phi)*np.sin(Lambda)
Z = RAD*np.cos(Lambda)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection="3d")

ax.scatter(X, Y, Z, alpha=0.2, c=Z)

ax.set_xlim((-1.2*RAD, 1.2*RAD))
ax.set_ylim((-1.2*RAD, 1.2*RAD))
ax.set_zlim((-1.2*RAD, 1.2*RAD))

ax.set_xlabel("x")
ax.set_ylabel("y")

ax.set_title(r"{:d} uniformly distributed random vectors on a sphere".format(NUM_POINTS_3D))

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11,5))

# top view
ax1.scatter(X, Y, c=Z, alpha=0.2, s=10)

ax1.set_xlim((-1.2*RAD, 1.2*RAD))
ax1.set_ylim((-1.2*RAD, 1.2*RAD))

ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax1.set_title(r"Top view")

# side view
ax2.scatter(X, Z, c=Z, alpha=0.2, s=10)

ax2.set_xlim((-1.2*RAD, 1.2*RAD))
ax2.set_ylim((-1.2*RAD, 1.2*RAD))

ax2.set_xlabel("x")
ax2.set_ylabel("z")

ax2.set_title(r"Side view")

plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,3))

# x-components
ax1.hist(X, bins=np.linspace(-1, 1, 100), color="tab:green")

ax1.set_xlabel("x value")
ax1.set_ylabel("number of occurences")

# y-components
ax2.hist(Y, bins=np.linspace(-1, 1, 100), color="tab:green")

ax2.set_xlabel("y value")
ax2.set_ylabel("number of occurences")

# z-components
ax3.hist(Z, bins=np.linspace(-1, 1, 100), color="tab:green")

ax3.set_xlabel("z value")
ax3.set_ylabel("number of occurences")

# ax.set_title("histogram")

plt.show()
```
